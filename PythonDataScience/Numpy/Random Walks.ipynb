### Random Walks

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
import random
```

#### Creamos un trayecto aleatorio de 1000 pasos

```python
position = 0
walk= [position]
steps= 1000

for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
    
plt.plot(walk[:100])
```

#### Puede hacerse utilizando un array

```python
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps) #  1000 pasos aleatorios
steps = np.where(draws > 0, 1, -1) # condicionamos los valores + y -
walk = steps.cumsum() # acumula la suma de los valores desde el inicio

plt.plot(walk[:100]) # dibujamos los 100 primeros pasos
```

```python
(np.abs(walk) >= 10).argmax()
```
