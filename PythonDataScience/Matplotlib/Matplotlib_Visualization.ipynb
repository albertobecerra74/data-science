### Simple Line Plots

La más sencilla de todas las representaciones es dibujar una función simple y = f(x)

```python
plt.style.use('seaborn-whitegrid')
```

```python
# creamos un array
x = np.linspace(0, 10, 1000)
```

Para todos los plots generados a través de Matplotlib, empezaremos creando una "figure" y unos "axes"

#### Figure

Es el contenedor donde vamos a representar el gráfico. Contiene todos los objetos que forman parte de este: axes, graphics, text and labels

#### Axes

Lo que vemos por encima del contenedor (figure), una caja que contiene las etiquetas y los gráficos que vemos

```python
# figure
fig = plt.figure()
# axes
ax = plt.axes()
# gráfico
ax.plot(x, np.sin(x))
```

Podemos hacer uso del interfaz pylab, para hacerlo más sencillo, que internamente crea la "figure" y los "axes"

```python
# imprimimos la función seno del array
plt.plot(x, np.sin(x))
```

```python
# Si queremos imprimir varias figuras simples en el mismo dibujo
# simplemente, llamamos a la función plot varias veces
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

#### Ajustando el color de las líneas y el estilo

Podemos personalizar el color de cada una de las líneas, así como el tipo de línea que usamos para representarlas. Para ello utilizamos los argumentos:

* color : acepta multitud de formas de especificar los colores
    - 'blue'
    - 'g'
    - '0.75'
    - '#FFDD44)
    - etc
    
* linestyle
    - '-' solid
    - '--' dashed
    - '-.' dashdot
    - ':' dotted
    

```python
# definimos el color y el formato de la línea
plt.plot(x, np.sin(x), color= 'red', linestyle='-')
plt.plot(x, np.cos(x), color= 'gray', linestyle='-.')
```

#### Ajustando el límite de los ejes

Matplotlib autoajusta los ejes a la información que representa. De todas formas, se pueden modificar estos usando los métodos:

* xlim()
* ylim()

En ambos, se especifican los límites superior e inferior

```python
plt.plot(x, np.sin(x))
plt.xlim= (10, 0)
plt.ylim= (-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis= ('equal')
```

```python
plt.plot(x, np.sin(x), label='sin(x)')

plt.title= ('A Sine Curve')
plt.xlabel = ('x')
plt.ylabel = ('sin(x)')
plt.xlim= (10, 0)
plt.ylim= (-1.5, 1.5)

```

```python
# utilizando ax methods, podemos representar todo lo anterior
ax = plt.axes()
ax.plot(x, np.sin(x), label='sin(x)')
ax.set(title='A Sine Curve', xlabel = 'x',
       ylabel = 'sin(x)', xlim= (-1, 11),
       ylim= (-1.5, 1.5))
```

### Simple Scatter Plots

En lugar de unir los puntos a través de una línea, aquí los puntos aparecen como tal, representados individualmente

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
```

```python
# creamos un array
x = np.linspace(0,10, 30)
```

```python
# calculamos la función seno
y = np.sin(x)
```

```python
# ploteamos la función ('o' es el tipo de marcador que 
# representa cada punto)
plt.plot(x,y, 'o', color='black')
```

```python
plt.plot(x, y, color='red')
```

```python
# ejemplo de tipos de marcadores distintos, ahí bastantes más que estos
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '<']:
            plt.plot(rng.rand(5),rng.rand(5), marker,
            label = "marker='{0}'".format(marker))
        
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

```python
# se puede plotear con la línea y los puntos
# '-ok' -> line (-), circle marker(o), black(k)
plt.plot(x, y, '-ok', color='red')
```

```python
# más ejemplos
plt.plot(x, y, '-p', color='gray',
        markersize=15, linewidth=4,
        markerfacecolor = 'white',
        markeredgecolor = 'gray',
        markeredgewidth = 2)

plt.ylim(-1.2,1.2)
```

### Scatter Plots with plt.scatter

Un segundo método más poderoso para crear plots de puntos es plt.scatter, que puede ser utilizada de forma bastante similar
a plt.plot

```python
plt.scatter(x, y, marker='o')
```

La principal diferencia es que con scatter podemos crear gráficos variando las propiedades de cada punto individual (tamaño, forma, color, transparencia, etc)

```python
rng = np.random.RandomState(0)

x= rng.randn(100)
y= rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x,y, c=colors, s=sizes, alpha=0.4, cmap='viridis')
plt.colorbar() #show color scale
```

El color y el tamaño puede ser utilizado para mostrar información adicional en la visualización, con el objetivo de reflejar datos multidimensionales

```python
# ejemplo
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
```

```python
features.shape
```

```python
features.ndim
```

```python
plt.scatter(features[0],features[1], c=iris.target, s=100*features[3], alpha=0.2, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

#### plot versus scatter

Para grandes datasets es más eficiente el uso de plot que de scatter. Esto es debido a que scatter, tiene la capacidad de renderizar un distinto tipo y color por cada punto, lo que supone un coste extra en el tiempo de construcción de cada uno de ellos; sin embargo, plot pinta los puntos siempre iguales, clones unos de otros, por lo que la apariencia de cada punto se copia al resto y se hace sólo una vez por cada uno de los puntos.


### Basic Errorbars

Un gráfico básico de barras de error, puede ser representado fácilmente con una línea de código. 

Para más información consultar plt.errorbar

```python
# creamos los datos
x= np.linspace(0, 10, 50)
dy = 0.8
y=np.sin(x) + dy * np.random.randn(50)

# dibujamos el gráfico
#  - el tamaño de la línea de error se define con yerr
#  - el formato de las líneas y puntos se hace igual que en plot
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
# podemos hacer las líneas de error más transparentes que los puntos
# para que se vea mejor la información
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
            ecolor='lightgray', elinewidth=3, capsize=0)
```
